```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import gensim
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Precily_Text_Similarity.csv")
```

```python
df
```

```python
filedocument=[]
for i in df['text1']:
    filedocument.append(i)
```

```python
filedocument
```

```python
gen_docs=[[w.lower() for  w in word_tokenize(text)] for text in filedocument]
```

```python
dictionary=gensim.corpora.Dictionary(gen_docs)
corpus=[dictionary.doc2bow(gen_doc) for gen_doc in gen_docs]
```

```python
tf_idf=gensim.models.TfidfModel(corpus) 
```

```python
sims=gensim.similarities.Similarity('templates/',tf_idf[corpus],num_features=len(dictionary))
```

```python
file2document=[]
for i in df['text2']:
    file2document.append(i)
```

```python
for line in file2document:
    query=[w.lower() for w in word_tokenize(line)]
    query_doc_bow=dictionary.doc2bow(query)
```

```python
query_doc_tf_idf=tf_idf[query_doc_bow]
sum_of_sims=(np.sum(sims[query_doc_tf_idf],dtype=np.float32))
```

```python
percentage_of_sim=round(float(sum_of_sims/len(filedocument))*100)
print("percentage: ",percentage_of_sim)
```
